```python
import numpy as np 
data = [0, 10, 20, 25, 27, 30, 43, 56, 68, 70]
mean = np.mean(data)
std = np.std(data)

# 標準化
z = (data - mean) / std
print('standardized data(z): {}'.format(z))
print('mean: {:.2f}'.format(np.mean(z)))
print('std: {}'.format(np.std(z)))
```

MEMO :
基本的にある統計量を扱うときは今後標準化された値を扱うと考えてよい

```python
from sklearn.preprocessing import StandardScaler
data = np.array([0, 10, 20, 25, 27, 30, 43, 56, 68, 70])
print('data shape: {}'.format(data.shape))
data = np.expand_dims(data, axis=1)
print('reshapesd data shape: {}'.format(data.shape))

# インスタンス作成
scaler = StandardScaler()
scaled = scaler.fit_transform(data)
print(scaled)
```
