```python
import numpy as np 
import matplotlib.pyplot as plt 
import seaborn as sns
%matplotlib inline 

weight =np.array([42, 46, 53, 56, 58, 61, 62, 63, 65, 67, 73])
height = np.array([138, 150, 152, 163, 164, 167, 165, 182, 180, 180, 183])
plt.scatter(height, weight)
plt.plot(np.mean(height), np.mean(weight), 'ro')
plt.xlabel('weight')
plt.xlabel('height')
```

```python
sns.regplot(x=height, y=weight)
plt.plot(np.mean(height), np.mean(weight), 'ro')
plt.label('height')
plt.ylabel('weight')
```

```python
from sklearn.linear_model import LinearRegression
"""
This script performs a simple linear regression using the height and weight data.

Imports:
    from sklearn.linear_model import LinearRegression: Import the LinearRegression class from scikit-learn.

Variables:
    X: A 2D numpy array created by expanding the dimensions of the height array.
    y: The weight array.

Operations:
    reg = LinearRegression(): Instantiate the LinearRegression model.
    reg.fit(X, y): Fit the linear regression model using the height (X) and weight (y) data.
    print('b={}'.format(reg.coef_)): Print the coefficient (slope) of the linear regression model.
    print('a={}'.format(reg.intercept_)): Print the intercept of the linear regression model.
"""
X = np.expand_dims(height, axis=-1)
y = weight
reg = LinearRegression()
reg.fit(X, y)

print('b={}'.format(reg.coef_))
print('a={}'.format(reg.intercept_))
```

```python
# 標準偏差
s_x = np.std(height)
s_y = np.std(weight)
# 平均
mean_x = np.mean(height)
mean_y = np.mean(weight)
# 相関係数
r = np.corrcoef(weight, height)[0][1]
b = r * s_y / s_x
a = mean_y - b * mean_x
print('b={}'.format(b)) # sklearnの結果と一致
print('a={}'.format(a)) # sklearnの結果と一致
```

```python
# x軸の値作成
x = np.arange(130, 190, 1)
plt.plot(x, a + b * x)
# seabornのregplotも重ねて表示
sns.regplot(x=height, y=weight)
plt.xlabel('height')
plt.ylabel('weight')
```
